```python
# importing required libraries
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

### Loading the Dataset

```python
# reading the csv file and assigning it to a variable
energy_data = pd.read_csv("datasets/energydata_complete.csv")
```

```python
# loading the dataset
energy_data.head()
```

```python
# getting information about the dataset
energy_data.info()
```

```python
# getting a description of the dataset
energy_data.describe()
```

```python
# checking for empty fields in the dataset
energy_data.isnull().sum()
```

since there is no empty field in the dataset, we proceed further

```python
# dropping the date and lights columns
new_df = energy_data.drop(["date", "lights"], axis = 1)
```

```python
new_df
```

```python
new_df.info()
```

### Scaling

```python
# carrying out scaling on the dataset before splitting
# applying the Min Max Scaler technique using Fit Transform.

# retrieving just the numeric input values
data = new_df.values[:, :]

# performing a robust scaler transform of the dataset
scaler = MinMaxScaler()
data = scaler.fit_transform(data)

# convering the array back to a dataframe
scaled_data = pd.DataFrame(data, columns = new_df.columns)
```

```python
# loading the scaled dataset
scaled_data
```

We can see that the distributions have been adjusted and that the minimum and maximum values for each variable are now a crisp 0.0 and 1.0 respectively.



```python
# summarizing the dataset
scaled_data.describe()
```

# Linear regression on T2 and T6

### Training & Testing Data

```python
# defining the dependent and independent variables
X = scaled_data.loc[:, ["T2"]].values
y = scaled_data.loc[:, ["T6"]].values
```

```python
# splitting the scaled_dataset into the Training set and Test set using train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# fitting our Multiple Linear Regression to the Training set
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
# predicting the Test set results
y_preds = linear_reg.predict(X_test)
```

```python
y_preds
```

```python
# printing the intercept of the trained model
print('Intercept: \n', linear_reg.intercept_)
```

```python
# printing the coefficients of the trained model
print('Coefficients: \n', linear_reg.coef_)
```

```python
# R^2 value in two D.P
round(r2_score(y_test, y_preds), 2)
```

# Linear Regression on entire dataset

```python
# defining the dependent and independent variables
X = scaled_data.drop('Appliances', 1)
y = scaled_data['Appliances']
```

```python
# splitting the scaled_dataset into the Training set and Test set using train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
#checking for shape of data split
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# fitting our Multiple Linear Regression to the Training set
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
# predicting the Test set results
y_preds = linear_reg.predict(X_test)
```

```python
y_preds
```

```python
# calculating the mean absolute error (MAE) in 2 D.P
mae = mean_absolute_error(y_test, y_preds)
print("The Mean Absolute Error (MAE) is %.2f" %mae)
```

```python
rss = np.square((y_test - y_preds)).sum()
print('The Residual Sum of Squares (RSS) is %.2f'%rss)
```

```python
# calculating the root mean squared error in 3 D.P
rmse = np.sqrt(mean_squared_error(y_test, y_preds))
print("The Root Mean Squared Error (RMSE) is %.3f" %rmse)
```

```python
# coefficient of determination in 2 D.P
r2 = r2_score(y_test, y_preds)
print("The Coefficient of Determination is %.2f" %r2)
```

```python
# printing the coefficients of the trained model
linear_reg_coef = linear_reg.coef_
feat_imp = pd.DataFrame(linear_reg_coef, index=X.columns, columns=['Coefficients'])
#feat_imp['absolute coeff'] = feat_imp['Coefficients'].apply(np.abs)
feat_imp.sort_values('Coefficients')
```

# Ridge on entire dataset

```python
# training with a Ridge model with an alpha value of 0.4
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(X_train, y_train)
```

```python
y_preds = ridge_reg.predict(X_test)
```

```python
# calculating the root mean squared error in 3 D.P
rmse = np.sqrt(mean_squared_error(y_test, y_preds))
print("The Root Mean Squared Error (RMSE) is %.3f" %rmse)
```

# LASSO on entire dataset

```python
# training with a Lassso model with an alpha value of 0.001
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(X_train, y_train)
```

```python
y_preds = lasso_reg.predict(X_test)
```

```python
# printing the coefficients of the trained model
lasso_reg_coef = lasso_reg.coef_
feat_imp = pd.DataFrame(lasso_reg_coef, index=X.columns, columns=['Coefficients'])
feat_imp['absolute coeff'] = feat_imp['Coefficients'].apply(np.abs)
feat_imp.sort_values('absolute coeff', ascending=False)
```

```python
# calculating the root mean squared error in 3 D.P
rmse = np.sqrt(mean_squared_error(y_test, y_preds))
print("The Root Mean Squared Error (RMSE) is %.3f" %rmse)
```
